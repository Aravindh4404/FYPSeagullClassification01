```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split
import numpy as np
import random
import matplotlib.pyplot as plt
from datetime import datetime
import timm  # For pretrained ViT models
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

# For attention visualization
import cv2
from PIL import Image

# ----------------------------
# 1. Setup & Reproducibility
# ----------------------------
torch.manual_seed(42)
np.random.seed(42)
random.seed(42)

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Define a folder to save checkpoints using the current date
date_str = datetime.now().strftime('%Y%m%d')
checkpoint_folder = f'/content/drive/My Drive/FYP/InterpretableViT_{date_str}/'
os.makedirs(checkpoint_folder, exist_ok=True)

# ----------------------------
# 2. Data Transforms & Datasets
# ----------------------------
transform_train = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.5,0.5,0.5],[0.5,0.5,0.5]),
])

transform_val_test = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize([0.5,0.5,0.5],[0.5,0.5,0.5]),
])

# Paths for training and test datasets
data_path = '/content/drive/My Drive/FYP/Dataset/HQ3/train'
test_data_path = '/content/drive/My Drive/FYP/Dataset/HQ3/test'

# Create datasets
full_train_dataset = datasets.ImageFolder(data_path, transform=transform_train)
test_dataset = datasets.ImageFolder(test_data_path, transform=transform_val_test)

# Split the training dataset: 80% train, 20% validation
train_size = int(0.95 * len(full_train_dataset))
val_size = len(full_train_dataset) - train_size
train_dataset, val_dataset = random_split(full_train_dataset, [train_size, val_size])

batch_size = 16
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=4, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=4, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=4, pin_memory=True)

# ----------------------------
# 3. Improved & Interpretable ViT Architecture
# ----------------------------
class InterpretableViT(nn.Module):
    def __init__(self, dropout_rate=0.3, hidden_dim=512):
        """
        This model uses a pre-trained ViT backbone and removes its original classification head.
        It then:
          - Extracts the [CLS] token (which is already well-trained)
          - Computes a learned attention over patch tokens (the remaining tokens)
          - Aggregates the patch tokens via a weighted sum
          - Concatenates the CLS token and the weighted patch summary
          - Feeds the combined representation through a custom MLP classifier
        The attention weights are returned for later visualization.
        """
        super(InterpretableViT, self).__init__()
        self.vit = timm.create_model('vit_base_patch16_224', pretrained=True)
        self.vit.head = nn.Identity()  # Remove the original classification head

        # Determine embedding dimension (most ViT models have an attribute 'embed_dim')
        self.embed_dim = self.vit.embed_dim if hasattr(self.vit, 'embed_dim') else 768

        # Attention layer on patch tokens (ignoring the CLS token)
        self.attention_layer = nn.Sequential(
            nn.Linear(self.embed_dim, 1)
        )

        # Classifier head: use both the CLS token and a weighted average of patch tokens.
        self.classifier = nn.Sequential(
            nn.LayerNorm(self.embed_dim * 2),
            nn.Dropout(dropout_rate),
            nn.Linear(self.embed_dim * 2, hidden_dim),
            nn.ReLU(),
            nn.Dropout(dropout_rate),
            nn.Linear(hidden_dim, 2)
        )

    def forward(self, x):
        # Get token embeddings from ViT; output shape: (B, N+1, embed_dim)
        tokens = self.vit.forward_features(x)
        # The first token is the [CLS] token
        cls_token = tokens[:, 0, :]  # Shape: (B, embed_dim)
        # Remaining tokens are patch tokens
        patch_tokens = tokens[:, 1:, :]  # Shape: (B, N, embed_dim)
        # Compute attention scores over patch tokens
        attn_scores = self.attention_layer(patch_tokens)  # (B, N, 1)
        attn_weights = torch.softmax(attn_scores, dim=1)    # (B, N, 1)
        # Weighted average of patch tokens
        weighted_patch = torch.sum(attn_weights * patch_tokens, dim=1)  # (B, embed_dim)
        # Combine the CLS token and weighted patch representation
        combined = torch.cat([cls_token, weighted_patch], dim=1)  # (B, 2*embed_dim)
        logits = self.classifier(combined)  # (B, 2)
        return logits, attn_weights  # Return logits and attention weights for interpretability

# Initialize the model and send it to the device
model = InterpretableViT().to(device)

# ----------------------------
# 4. Loss, Optimizer, and Scheduler
# ----------------------------
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.parameters(), lr=0.0001, weight_decay=1e-4)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.1, patience=3, verbose=True)

# ----------------------------
# 5. Training, Validation, and Test Functions
# ----------------------------
def train(model, train_loader, val_loader, criterion, optimizer, scheduler, epochs=20):
    best_acc = 0.0
    train_losses = []
    val_accuracies, val_precisions, val_recalls, val_f1s = [], [], [], []

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            logits, _ = model(inputs)
            loss = criterion(logits, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

        train_loss = running_loss / len(train_loader)
        train_losses.append(train_loss)
        print(f"Epoch [{epoch+1}/{epochs}] Training Loss: {train_loss:.6f}")

        # Validate
        val_acc, val_precision, val_recall, val_f1 = validate(model, val_loader, criterion)
        val_accuracies.append(val_acc)
        val_precisions.append(val_precision)
        val_recalls.append(val_recall)
        val_f1s.append(val_f1)

        scheduler.step(val_acc)

        # Save the model every epoch
        final_model_path = os.path.join(checkpoint_folder, f"final_model_vit_{date_str}.pth")
        torch.save(model.state_dict(), final_model_path)
        print(f"Final model (epoch {epoch+1}) saved at {final_model_path}")

        # Save the best model so far
        if val_acc > best_acc:
            best_acc = val_acc
            best_model_path = os.path.join(checkpoint_folder, f"best_model_vit_{date_str}.pth")
            torch.save(model.state_dict(), best_model_path)
            print(f"New best model with accuracy: {best_acc:.2f}% saved at {best_model_path}")

    # Plot metrics
    plt.figure(figsize=(20, 10))
    plt.subplot(2, 2, 1)
    plt.plot(train_losses, label='Training Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training Loss')
    plt.grid(True)

    plt.subplot(2, 2, 2)
    plt.plot(val_accuracies, label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy (%)')
    plt.title('Validation Accuracy')
    plt.grid(True)

    plt.subplot(2, 2, 3)
    plt.plot(val_precisions, label='Validation Precision')
    plt.xlabel('Epochs')
    plt.ylabel('Precision')
    plt.title('Validation Precision')
    plt.grid(True)

    plt.subplot(2, 2, 4)
    plt.plot(val_recalls, label='Validation Recall')
    plt.xlabel('Epochs')
    plt.ylabel('Recall')
    plt.title('Validation Recall')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(10, 5))
    plt.plot(val_f1s, label='Validation F1 Score')
    plt.xlabel('Epochs')
    plt.ylabel('F1 Score')
    plt.title('Validation F1 Score')
    plt.grid(True)
    plt.show()

    print("Training complete!")

def validate(model, loader, criterion):
    model.eval()
    correct = 0
    total = 0
    all_labels = []
    all_predictions = []

    with torch.no_grad():
        for inputs, labels in loader:
            inputs, labels = inputs.to(device), labels.to(device)
            logits, _ = model(inputs)
            _, predicted = torch.max(logits.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            all_labels.extend(labels.cpu().numpy())
            all_predictions.extend(predicted.cpu().numpy())

    accuracy = 100 * correct / total
    precision = precision_score(all_labels, all_predictions, average='binary', zero_division=0)
    recall = recall_score(all_labels, all_predictions, average='binary', zero_division=0)
    f1 = f1_score(all_labels, all_predictions, average='binary', zero_division=0)
    print(f'Validation Accuracy: {accuracy:.2f}%, Precision: {precision:.2f}, Recall: {recall:.2f}, F1 Score: {f1:.2f}')
    return accuracy, precision, recall, f1

def test(model, loader, criterion):
    model.eval()
    test_loss = 0.0
    correct = 0
    total = 0

    with torch.no_grad():
        for inputs, labels in loader:
            inputs, labels = inputs.to(device), labels.to(device)
            logits, _ = model(inputs)
            loss = criterion(logits, labels)
            test_loss += loss.item()
            _, predicted = torch.max(logits.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    accuracy = 100 * correct / total
    print(f'Test Loss: {test_loss/len(loader):.6f}, Test Accuracy: {accuracy:.2f}%')

# ----------------------------
# 6. Attention Visualization Function
# ----------------------------
def visualize_attention(image_path, attn_weights):
    """
    Given an image path and the attention weights from the model (for one sample),
    this function visualizes:
      - The original image
      - The raw attention map (reshaped to 14x14, since ViT base has 196 patch tokens)
      - An overlay of the heatmap on the original image
    """
    # Load original image and resize to 224x224
    image = np.array(Image.open(image_path).convert("RGB").resize((224, 224)))

    # attn_weights shape: (B, N, 1), assume B = 1 and N = 196 (14x14 patches)
    attn_map = attn_weights[0].squeeze(-1)  # shape: (196,)
    attn_map = attn_map.reshape(14, 14)

    # Normalize attention map
    attn_map = (attn_map - attn_map.min()) / (attn_map.max() - attn_map.min() + 1e-8)

    # Resize attention map to match original image size (224x224)
    attn_map_resized = cv2.resize(attn_map, (224, 224))

    # Generate a heatmap using OpenCV
    heatmap = cv2.applyColorMap(np.uint8(255 * attn_map_resized), cv2.COLORMAP_JET)
    heatmap = cv2.cvtColor(heatmap, cv2.COLOR_BGR2RGB)

    # Overlay the heatmap on the original image
    overlay = cv2.addWeighted(image, 0.6, heatmap, 0.4, 0)

    # Plot the results
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 3, 1)
    plt.imshow(image)
    plt.title("Original")
    plt.axis("off")

    plt.subplot(1, 3, 2)
    plt.imshow(attn_map_resized, cmap='jet')
    plt.title("Attention Map")
    plt.axis("off")

    plt.subplot(1, 3, 3)
    plt.imshow(overlay)
    plt.title("Overlay")
    plt.axis("off")

    plt.show()

# ----------------------------
# 7. Main Training & Evaluation
# ----------------------------
train(model, train_loader, val_loader, criterion, optimizer, scheduler, epochs=20)
test(model, test_loader, criterion)
```

```python
!pip install captum
!pip install timm
import torch
import torch.nn as nn
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import timm
import os
from typing import Tuple, Optional, Dict, List
from torchvision import transforms
from captum.attr import FeatureAblation
from datetime import datetime
import pandas as pd
from tqdm import tqdm
from pathlib import Path
import random
import warnings
import time

# Suppress future warnings about torch.load
warnings.filterwarnings('ignore', category=FutureWarning)

class ViTModified(nn.Module):
    """Previous ViTModified class implementation remains the same"""
    def __init__(self, num_classes: int = 2, pretrained: bool = True):
        super().__init__()
        self.vit = timm.create_model('vit_base_patch16_224', pretrained=pretrained)
        num_ftrs = self.vit.head.in_features
        self.vit.head = nn.Linear(num_ftrs, num_classes)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.vit(x)

class EnhancedFeatureAblationAnalyzer:
    def __init__(
        self,
        model: nn.Module,
        class_names: list,
        save_dir: str,
        device: str = 'cuda' if torch.cuda.is_available() else 'cpu',
        patch_size: int = 16,
        confidence_threshold: float = 0.0,
        max_images_per_class: int = None,
        batch_size: int = 4  # Added batch processing
    ):
        """
        Enhanced analyzer with batch processing and image limits.

        Args:
            model: The neural network model
            class_names: List of class names
            save_dir: Directory to save results
            device: Computing device (cuda/cpu)
            patch_size: Size of patches for ablation
            confidence_threshold: Minimum confidence score to save results
            max_images_per_class: Maximum number of images to process per class
            batch_size: Number of images to process in parallel
        """
        self.model = model.to(device)
        self.device = device
        self.class_names = class_names
        self.patch_size = patch_size
        self.confidence_threshold = confidence_threshold
        self.max_images_per_class = max_images_per_class
        self.batch_size = batch_size

        # Create directory structure
        self.base_dir = Path(save_dir)
        self.results_dir = self.base_dir / datetime.now().strftime("%Y%m%d_%H%M%S")
        self.viz_dir = self.results_dir / "visualizations"
        self.stats_dir = self.results_dir / "statistics"
        self.create_directories()

        # Initialize transforms
        self.transform = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5]),
        ])

        # Initialize Feature Ablation
        self.feature_ablator = FeatureAblation(self.model)

        # Analysis results storage
        self.results_data = []

    def create_directories(self):
        """Previous create_directories implementation remains the same"""
        self.results_dir.mkdir(parents=True, exist_ok=True)
        self.viz_dir.mkdir(parents=True, exist_ok=True)
        self.stats_dir.mkdir(parents=True, exist_ok=True)

    def get_image_paths(self, class_dir: Path) -> List[Path]:
        """Get a list of image paths, respecting the maximum images limit."""
        image_paths = list(class_dir.glob("*.[jp][pn][gf]"))
        if self.max_images_per_class is not None:
            random.shuffle(image_paths)  # Randomize selection
            image_paths = image_paths[:self.max_images_per_class]
        return image_paths

    def process_batch(self, image_paths: List[str]) -> List[Dict]:
        """Process a batch of images simultaneously."""
        results = []
        batch_tensors = []
        valid_paths = []

        # Prepare batch
        for img_path in image_paths:
            try:
                image = Image.open(img_path).convert("RGB")
                image_resized = image.resize((224, 224))
                image_tensor = self.transform(image_resized).unsqueeze(0)
                batch_tensors.append(image_tensor)
                valid_paths.append(img_path)
            except Exception as e:
                print(f"Error loading {img_path}: {str(e)}")
                continue

        if not batch_tensors:
            return results

        # Stack tensors into a single batch
        batch = torch.cat(batch_tensors).to(self.device)

        # Get model predictions
        self.model.eval()
        with torch.no_grad():
            outputs = self.model(batch)
            probabilities = torch.nn.functional.softmax(outputs, dim=1)
            confidences = torch.max(probabilities, dim=1)
            pred_classes = outputs.argmax(dim=1)

        # Generate attribution maps for batch
        patch_mask = self.generate_patch_feature_mask()
        patch_mask = patch_mask.repeat(len(batch_tensors), 1, 1)

        attributions = self.feature_ablator.attribute(
            inputs=batch,
            baselines=torch.zeros_like(batch),
            target=pred_classes.tolist(),
            feature_mask=patch_mask
        )

        # Process results
        for idx, (img_path, conf, pred_class) in enumerate(zip(valid_paths, confidences[0], pred_classes)):
            try:
                # Process attribution map
                attr_map = np.mean(np.abs(attributions[idx].cpu().numpy()), axis=0)
                attr_map = (attr_map - attr_map.min()) / (attr_map.max() - attr_map.min() + 1e-8)
                attr_map = cv2.GaussianBlur(attr_map, (3, 3), 0)

                # Save visualization if confidence threshold is met
                if conf >= self.confidence_threshold:
                    image = Image.open(img_path).convert("RGB")
                    image_resized = np.array(image.resize((224, 224)))
                    self.save_visualization(
                        image_resized,
                        attr_map,
                        img_path,
                        pred_class.item(),
                        conf.item()
                    )

                results.append({
                    'image_path': str(img_path),
                    'predicted_class': self.class_names[pred_class.item()],
                    'confidence': conf.item(),
                    'original_size': Image.open(img_path).size,
                    'attribution_stats': {
                        'mean': float(attr_map.mean()),
                        'std': float(attr_map.std()),
                        'max': float(attr_map.max()),
                        'min': float(attr_map.min())
                    }
                })

            except Exception as e:
                print(f"Error processing results for {img_path}: {str(e)}")
                continue

        return results

    def analyze_dataset(self, data_dir: str):
        """
        Analyze images in the specified directory structure with progress tracking.
        """
        data_dir = Path(data_dir)
        all_results = []

        # Process each class directory
        for class_dir in data_dir.iterdir():
            if not class_dir.is_dir():
                continue

            print(f"\nProcessing {class_dir.name}...")
            image_paths = self.get_image_paths(class_dir)

            # Process images in batches with progress bar
            for i in tqdm(range(0, len(image_paths), self.batch_size)):
                batch_paths = image_paths[i:i + self.batch_size]
                results = self.process_batch(batch_paths)

                for result in results:
                    result['true_class'] = class_dir.name
                    all_results.append(result)

                # Save intermediate results every 50 images
                if len(all_results) % 50 == 0:
                    self.save_analysis_results(all_results)

        # Save final results
        self.save_analysis_results(all_results)

    # Other methods remain the same...

def main():
    # Model setup with weights_only=True for security
    model = ViTModified()
    model.load_state_dict(torch.load(
        '/content/drive/My Drive/FYP/VIT2_HQ2_20241224/best_model_vit_20241224.pth',
        map_location='cuda' if torch.cuda.is_available() else 'cpu',
        weights_only=True  # Added for security
    ))

    # Initialize analyzer with limits and batch processing
    analyzer = EnhancedFeatureAblationAnalyzer(
        model=model,
        class_names=['Glaucous_Winged_Gull', 'Slaty_Backed_Gull'],
        save_dir='/content/drive/My Drive/FYP/VITFeatureAblation',
        confidence_threshold=0.5,
        max_images_per_class=150,  # Process maximum 50 images per class
        batch_size=4  # Process 4 images at once
    )

    # Process dataset
    data_dir = '/content/drive/My Drive/FYP/Entire Bird Segment'
    analyzer.analyze_dataset(data_dir)

if __name__ == "__main__":
    main()
```

```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split
import numpy as np
import random
import matplotlib.pyplot as plt
from datetime import datetime
import timm  # For pretrained Vision Transformer models
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

# ----------------------------
# 1. Reproducibility & Setup
# ----------------------------
torch.manual_seed(42)
np.random.seed(42)
random.seed(42)

# Mount Google Drive (if using Google Colab)
from google.colab import drive
drive.mount('/content/drive')

# Define folder to save model checkpoints (using current date)
date_str = datetime.now().strftime('%Y%m%d')
checkpoint_folder = f'/content/drive/My Drive/FYP/VIT2_HQ2_{date_str}/'
os.makedirs(checkpoint_folder, exist_ok=True)

# ----------------------------
# 2. Data Augmentation & Transforms
# ----------------------------
# Use RandAugment to further enrich your training set
transform_train = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandAugment(num_ops=2, magnitude=9),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5]),
])

# For validation and testing, only basic resizing and normalization
transform_val_test = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5]),
])

# ----------------------------
# 3. Datasets & DataLoaders
# ----------------------------
data_path = '/content/drive/My Drive/FYP/Dataset/HQ3/train'
test_data_path = '/content/drive/My Drive/FYP/Dataset/HQ3/test'
full_train_dataset = datasets.ImageFolder(data_path, transform=transform_train)
test_dataset = datasets.ImageFolder(test_data_path, transform=transform_val_test)

# Split full training dataset: 80% train, 20% validation
train_size = int(0.95 * len(full_train_dataset))
val_size = len(full_train_dataset) - train_size
train_dataset, val_dataset = random_split(full_train_dataset, [train_size, val_size])

batch_size = 16
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=4, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=4, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=4, pin_memory=True)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ----------------------------
# 4. Model Definition
# ----------------------------
class ViTModified(nn.Module):
    def __init__(self):
        super(ViTModified, self).__init__()
        # Load a pretrained ViT from timm
        self.vit = timm.create_model('vit_base_patch16_224', pretrained=True)
        num_ftrs = self.vit.head.in_features
        # Replace the classification head for binary classification
        self.vit.head = nn.Linear(num_ftrs, 2)

    def forward(self, x):
        return self.vit(x)

model = ViTModified().to(device)

# ----------------------------
# 5. Loss, Optimizer, and Scheduler
# ----------------------------
criterion = nn.CrossEntropyLoss()

# Use AdamW optimizer and a higher max_lr for OneCycleLR
optimizer = optim.AdamW(model.parameters(), lr=1e-3, weight_decay=1e-4)

# OneCycleLR scheduler: note that steps_per_epoch is the number of batches in train_loader
epochs = 25
scheduler = torch.optim.lr_scheduler.OneCycleLR(
    optimizer, max_lr=1e-3, steps_per_epoch=len(train_loader), epochs=epochs
)

# For mixed precision training
scaler = torch.cuda.amp.GradScaler()

# ----------------------------
# 6. Training Loop
# ----------------------------
def train(model, train_loader, val_loader, criterion, optimizer, scheduler, epochs):
    best_acc = 0.0
    train_losses, val_accuracies, val_precisions, val_recalls, val_f1s = [], [], [], [], []

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()

            # Use mixed precision autocast
            with torch.cuda.amp.autocast():
                outputs = model(inputs)
                loss = criterion(outputs, labels)

            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            running_loss += loss.item()
            scheduler.step()  # OneCycleLR updates every batch

        train_loss = running_loss / len(train_loader)
        train_losses.append(train_loss)
        print(f"Epoch [{epoch+1}/{epochs}] Training Loss: {train_loss:.6f}")

        # Validate and record metrics
        val_acc, val_precision, val_recall, val_f1 = validate(model, val_loader, criterion)
        val_accuracies.append(val_acc)
        val_precisions.append(val_precision)
        val_recalls.append(val_recall)
        val_f1s.append(val_f1)

        # Save model checkpoint every epoch ("final" model)
        final_model_path = os.path.join(checkpoint_folder, f"final_model_vit_{date_str}.pth")
        torch.save(model.state_dict(), final_model_path)
        print(f"Final model (epoch {epoch+1}) saved at {final_model_path}")

        # Save the best model so far
        if val_acc > best_acc:
            best_acc = val_acc
            best_model_path = os.path.join(checkpoint_folder, f"best_model_vit_{date_str}.pth")
            torch.save(model.state_dict(), best_model_path)
            print(f"New best model with accuracy: {best_acc:.2f}% saved at {best_model_path}")

    # Plot metrics
    plt.figure(figsize=(20, 10))
    plt.subplot(2, 2, 1)
    plt.plot(train_losses, label='Training Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training Loss')
    plt.grid(True)

    plt.subplot(2, 2, 2)
    plt.plot(val_accuracies, label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy (%)')
    plt.title('Validation Accuracy')
    plt.grid(True)

    plt.subplot(2, 2, 3)
    plt.plot(val_precisions, label='Validation Precision')
    plt.xlabel('Epochs')
    plt.ylabel('Precision')
    plt.title('Validation Precision')
    plt.grid(True)

    plt.subplot(2, 2, 4)
    plt.plot(val_recalls, label='Validation Recall')
    plt.xlabel('Epochs')
    plt.ylabel('Recall')
    plt.title('Validation Recall')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(10, 5))
    plt.plot(val_f1s, label='Validation F1 Score')
    plt.xlabel('Epochs')
    plt.ylabel('F1 Score')
    plt.title('Validation F1 Score')
    plt.grid(True)
    plt.show()

    print("Training complete!")

# ----------------------------
# 7. Validation Function
# ----------------------------
def validate(model, loader, criterion):
    model.eval()
    correct, total = 0, 0
    all_labels, all_predictions = [], []

    with torch.no_grad():
        for inputs, labels in loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            all_labels.extend(labels.cpu().numpy())
            all_predictions.extend(predicted.cpu().numpy())

    accuracy = 100 * correct / total
    precision = precision_score(all_labels, all_predictions, average='binary', zero_division=0)
    recall = recall_score(all_labels, all_predictions, average='binary', zero_division=0)
    f1 = f1_score(all_labels, all_predictions, average='binary', zero_division=0)
    print(f'Validation Accuracy: {accuracy:.2f}%, Precision: {precision:.2f}, Recall: {recall:.2f}, F1 Score: {f1:.2f}')
    return accuracy, precision, recall, f1

# ----------------------------
# 8. Test Function
# ----------------------------
def test(model, loader, criterion):
    model.eval()
    test_loss, correct, total = 0.0, 0, 0

    with torch.no_grad():
        for inputs, labels in loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            test_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    accuracy = 100 * correct / total
    print(f'Test Loss: {test_loss/len(loader):.6f}, Test Accuracy: {accuracy:.2f}%')

# ----------------------------
# 9. Main Training & Evaluation
# ----------------------------
train(model, train_loader, val_loader, criterion, optimizer, scheduler, epochs=epochs)
test(model, test_loader, criterion)
```
